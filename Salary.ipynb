```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = df[['index', 'Job Title', 'Rating', 'Company Name', 'Location', 'Lower Salary', 'Upper Salary',
       'Avg Salary(K)']]
```

```python
df1.dropna()
```

```python
num_vagas = dict(df1['Job Title'].value_counts())
```

```python
num_vagas
```

```python
remover_itens = []

for chave, valor in num_vagas.items():
    if num_vagas[chave] <= 3:
        remover_itens.append(chave)

for item in remover_itens:
    num_vagas.pop(item)

print(num_vagas)
```

```python
quantidade = list(num_vagas.values())
vagas = list(num_vagas.keys())

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(vagas, quantidade)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Show Plot
plt.show()
```

```python
df_tools = df[['Python','spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit',
              'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an']]
```

```python
df_tools
```

```python
dicionario = {}
for i in df_tools.columns:
    dicionario[i] = df_tools[i].sum()
    print(f'Total de vagas que requerem {i}: {df_tools[i].sum()}')
```

```python
dicionario
```

```python
lista = []
lista2 = []
for i in sorted(dicionario, key = dicionario.get, reverse=True):
    lista.append(i)
    lista2.append(dicionario[i])
    print(f'{i}: {dicionario[i]}')
```

```python
#lista -> ferramenta (vagas)
#lista2 -> quantidade

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Vertical Bar Plot
ax.bar(lista, lista2, color='black', edgecolor='blue')

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 10)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = False, color ='black',
        linestyle ='-', linewidth = 0.3,
        alpha = 0.5)
 
# Add annotation to bars
rects = ax.patches    

for rect, label in zip(rects, lista2):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
            ha='center', va='bottom', color='blue')

# Show Plot
plt.show()
```
